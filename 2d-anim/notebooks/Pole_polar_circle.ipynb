Math Behind random_point_on_circle

A circle of radius r centered at the origin is defined by:
$x^2 + y^2 = r^2$

In polar coordinates:
$x = r\cos\theta$
$y = r\sin\theta$

As $\theta$ goes from $0$ to $2\pi$, the point $(x,y)$ traces the entire circle.

Choosing a random angle $\theta \sim U(0,2\pi)$ gives a random point on the boundary.

Proof that the point lies on the circle:
$(r\cos\theta)^2 + (r\sin\theta)^2 = r^2(\cos^2\theta + \sin^2\theta) = r^2$

======================================================================================

Any point on a line segment between two points $(x_0, y_0)$ and $(x_1, y_1)$ can be written using **linear interpolation**:

$x = (1 - t) * x_0 + t * x_1$\
$y = (1 - t) * y_0 + t * y_1$

where t $\in$ [0, 1]:
- t = 0 gives the start point $(x_0, y_0)$
- t = 1 gives the end point $(x_1, y_1)$
- 0 < t < 1 gives any point in between


```python
from random import uniform, randint
import numpy as np
import matplotlib.pyplot as plt


def setup_axes():
	fig, ax = plt.subplots(figsize=(6,6), dpi=300, facecolor='lightblue', edgecolor='k', linewidth=3)
	ax.set_xlim(-10,10)
	ax.set_ylim(-10,10)
	ax.axhline(0, color='k',linewidth=1)
	ax.axvline(0, color='k',linewidth=1)
	ax.grid(True)
	ax.set_title('Polar Circle')
	ax.set_xlabel('x')
	ax.set_ylabel('y')
	return fig, ax


def circle(axes, set_of_points_1):
	axes.plot(radius*np.cos(set_of_points_1), radius*np.sin(set_of_points_1), color='r')
	return None

def points_around_circle(axes, point_x0, point_y0, point_x1, point_y1, point_xr, point_yr):
	axes.plot(point_x0, point_y0, "kx",markersize = 5, linewidth = 2, label=f"Point ({point_x0:.2f},{point_y0:.2f})")
	axes.plot(point_x1, point_y1, 'rx',markersize = 5, label=f"Point ({point_x1:.2f},{point_y1:.2f})")
	ax.plot([point_x0, point_x1], [point_y0, point_y1], "g-", linewidth=2, label="Chord")      #first all [x_cor] then [y_cord]
	ax.plot(point_xr, point_yr, "bo", markersize=4, label="Random point on chord")
	axes.legend(loc='upper left', fontsize=8)

def random_point_on_circle(r):
    theta = uniform(0, 2*np.pi)   # random angle
    return r*np.cos(theta), r*np.sin(theta)


def random_point_on_chord(x0, y0, x1, y1):
    t = uniform(0, 1)
    x = (1-t) * x0 + t * x1
    y = (1-t) * y0 + t * y1
    return x, y



figure, ax = setup_axes()
radius = int(input("Radius of circle: "))
points_set = np.linspace(-np.pi, np.pi, 100)
x0,y0 = random_point_on_circle(radius)
x1,y1 = random_point_on_circle(radius)
xr,yr = random_point_on_chord(x0,y0,x1,y1)

circle(ax, points_set)
points_around_circle(ax, x0, y0, x1, y1, xr, yr)

```
